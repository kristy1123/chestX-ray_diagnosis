```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 파일 읽기
df = pd.read_csv('C:\project\lungX\CXR8\Data_Entry_processed_Final.csv')

# 2. 분석하고 싶은 컬럼 리스트

counts=df['Finding Labels'].value_counts()
target_columns = ['No Finding', 'Infiltration', 'Effusion', 'Atelectasis']

```

### csv 데이터 확인 및 분석

### 파이차트

```python
# 4. 파이차트 시각화
plt.figure(figsize=(10, 10))

# 색상 설정 (No Finding은 회색, 나머지는 파스텔톤)
colors = []
for label in counts.index:
    if label == 'No Finding':
        colors.append('lightgray')
    else:
        # 나머지 병명은 파스텔톤 색상 자동 부여 (인덱스 기반)
        colors.append(plt.cm.Pastel1(len(colors))) 

plt.pie(counts, 
        labels=counts.index, 
        autopct='%1.1f%%',       # 비율 표시
        startangle=90,           # 12시 방향 시작
        colors=colors,           # 설정한 색상 적용
        wedgeprops={'edgecolor': 'white'}) # 깔끔한 테두리

plt.title("Distribution of Selected Finding Labels", fontsize=15)
plt.show()
```
